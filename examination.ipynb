```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
train_df = pd.read_csv('train.csv')
```

```python
train_df.columns
```

```python
#look at all columns of train df that begin with I


# M* - Market Dynamics/Technical features.
# E* - Macro Economic features.
# I* - Interest Rate features.
# P* - Price/Valuation features.
# V* - Volatility features.
# S* - Sentiment features.
# MOM* - Momentum features.
# D* - Dummy/Binary features.

for s in ['I', 'M', 'D', 'E', 'P', 'V', 'S', 'MOM']:
    print(train_df.filter(like=s).columns)
    plt.plot(train_df.filter(like=s))
    plt.legend(train_df.filter(like=s).columns)
    plt.show()
```
